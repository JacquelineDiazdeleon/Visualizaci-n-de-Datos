```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Datos de ejemplo
x = np.linspace(0, 10, 100)
y = np.sin(x)
```

```python
#Gráfico de lineas
plt.figure(figsize=(10, 4))
plt.plot(x, y)
plt.title('Gráfico de líneas')
plt.xlabel('X')
plt.ylabel('sin(x)')
plt.show()
```

```python
#Grafico de barras
categorias = ['A', 'B', 'C', 'D']
valores = [3, 7, 2, 5]
plt.figure(figsize=(10, 4))
plt.bar(categorias, valores)
plt.title('Gráfico de barras')
plt.xlabel('Categorías')
plt.ylabel('Valores')
plt.show()
```
